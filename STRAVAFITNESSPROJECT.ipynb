## ***2. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1: Average Steps by Day of Week(Daily Level)

```python
# Chart - 1 visualization code
# Ensure the column is datetime and reassign
daily_df['ActivityDate'] = pd.to_datetime(daily_df['ActivityDate'], errors='coerce')

# Now safely extract the day name
daily_df['Day'] = daily_df['ActivityDate'].dt.day_name()

# Plot average total steps by weekday
plt.figure(figsize=(10, 6))
sns.barplot(data=daily_df, x='Day', y='TotalSteps', estimator='mean', order=[
    'Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday'
])
plt.title('Average Total Steps by Day of Week')
plt.ylabel('Average Steps')
plt.xlabel('Day')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

##### 1. Why did you pick the specific chart?

This chart reveals daily behavior patterns ‚Äî are users more active on weekends or weekdays?

##### 2. What is/are the insight(s) found from the chart?

Users typically walk more on weekends, especially Saturday and Sunday, suggesting they are less active during workdays.

##### 3. Will the gained insights help creating a positive business impact?


This insight helps fitness app designers:

    Schedule push notifications, challenges, or reminders during low-activity weekdays.

    Promote weekend walking goals or social challenges to maintain momentum.

#### Chart - 2: Calories vs. Steps Scatter Plot(Daily Level)

```python
# Chart - 2 visualization code
plt.figure(figsize=(8, 6))
sns.scatterplot(data=daily_df, x='TotalSteps', y='Calories', hue='Day', palette='viridis')
plt.title('Calories vs Total Steps')
plt.xlabel('Total Steps')
plt.ylabel('Calories Burned')
plt.tight_layout()
plt.show()
```

##### 1. Why did you pick the specific chart?

To understand the relationship between calories burned and steps taken.

##### 2. What is/are the insight(s) found from the chart?

There is a positive correlation ‚Äî more steps generally lead to higher calories burned, though with some variation due to other factors (e.g., intensity or duration).

##### 3. Will the gained insights help creating a positive business impact?


Tailoring fitness goals based on user calorie/step targets.

Educating users that consistent movement (not necessarily intense workouts) drives calorie burn.

#### Chart - 3: Sedentary Minutes Distribution(Daily Level)

```python
# Chart - 3 visualization code
plt.figure(figsize=(8, 6))
sns.histplot(data=daily_df, x='SedentaryMinutes', bins=20, kde=True, color='coral')
plt.title('Distribution of Sedentary Minutes')
plt.xlabel('Sedentary Minutes')
plt.tight_layout()
plt.show()
```

##### 1. Why did you pick the specific chart?

Sedentary behavior is a critical risk factor for poor health. This chart shows how much time users remain inactive.

##### 2. What is/are the insight(s) found from the chart?

Many users have a high number of sedentary minutes, often well over 500‚Äì600 minutes (~8‚Äì10 hours per day).

##### 3. Will the gained insights help creating a positive business impact?


Reminders to stand or stretch regularly.

Alerts when prolonged inactivity is detected.

Development of personalized movement prompts.


#### Chart - 4: Average Calories Burnt by Hour of Day(Hourly Level Plot)

```python
# Chart - 4 visualization code
# Force datetime parsing again in case it's been dropped
hourly_df['ActivityHour'] = pd.to_datetime(hourly_df['ActivityHour'], errors='coerce')
hourly_df['Hour'] = hourly_df['ActivityHour'].dt.hour

# Prepare data for plotting
avg_calories_per_hour = hourly_df.groupby('Hour')['Calories'].mean().reset_index()

# Use non-interactive backend to prevent crashes
plt.figure(figsize=(10, 6))
sns.lineplot(data=avg_calories_per_hour, x='Hour', y='Calories', marker='o', color='teal')
plt.title('Average Calories Burned by Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Calories')
plt.grid(True)
plt.tight_layout()
plt.show()
```

##### 1. Why did you pick the specific chart?

To identify when users are most active (in terms of energy expenditure) during the day.

##### 2. What is/are the insight(s) found from the chart?

Peak calorie burn happens between morning and early evening hours, with a dip during early afternoon and night.

##### 3. Will the gained insights help creating a positive business impact?


Target time-based activity nudges (e.g., mid-day stretch reminders).

Optimize ad placement or motivational content during high/low activity windows.

#### Chart - 5: Boxplot of Calories by Hour(Hourly Level Plot)

```python
# Chart - 5 visualization code
plt.figure(figsize=(12, 6))
sns.boxplot(data=hourly_df, x='Hour', y='Calories')
plt.title('Calories Distribution by Hour')
plt.xlabel('Hour')
plt.ylabel('Calories')
plt.tight_layout()
plt.show()
```

##### 1. Why did you pick the specific chart?

To show variability in calorie burn across different hours ‚Äî not just averages, but also outliers and spread.

##### 2. What is/are the insight(s) found from the chart?

High variability during daytime hours, lower during sleep hours (late night to early morning).

##### 3. Will the gained insights help creating a positive business impact?


Supports:

    Personalized workout time recommendations (e.g., "You burn most at 6 PM").

    Segmented user targeting (e.g., night-shift workers vs day workers).

#### Chart - 6: Total Steps Per Minute (Distribution)(Minute Level Plot)

```python
# Chart - 6 visualization code

# Ensure datetime
minute_df['ActivityHour'] = pd.to_datetime(minute_df['ActivityHour'], errors='coerce')

# Plot distribution of TotalSteps
plt.figure(figsize=(10, 6))
sns.histplot(minute_df['TotalSteps'], bins=30, kde=True, color='dodgerblue')
plt.title('Distribution of Steps Per Minute')
plt.xlabel('Steps in One Minute')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()


```

##### 1. Why did you pick the specific chart?

To show how frequently users are moving intensely within 1-minute intervals.

##### 2. What is/are the insight(s) found from the chart?

Most 1-minute intervals show 0‚Äì20 steps, indicating light or sedentary behavior dominates.

##### 3. Will the gained insights help creating a positive business impact?


Helps create step streak or micro-goals (e.g., ‚ÄúTake 25 steps every 5 minutes‚Äù) to fight sedentary habits.

#### Chart - 7:  Total Steps by Hour (Minute-level Aggregation)(Minute Level Plot)

```python
# Chart - 7 visualization code
# Extract hour
minute_df['Hour'] = minute_df['ActivityHour'].dt.hour

# Sum steps per hour
steps_by_hour = minute_df.groupby('Hour')['TotalSteps'].sum().reset_index()

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(data=steps_by_hour, x='Hour', y='TotalSteps', palette='coolwarm')
plt.title('Total Steps by Hour (Aggregated from Minute-Level Data)')
plt.xlabel('Hour of Day')
plt.ylabel('Total Steps')
plt.tight_layout()
plt.show()
```

##### 1. Why did you pick the specific chart?

To understand when people accumulate the most steps ‚Äî granular pattern of movement.

##### 2. What is/are the insight(s) found from the chart?

Steps cluster during morning and evening (commute/exercise), with mid-day dips.

##### 3. Will the gained insights help creating a positive business impact?


Design timed interventions:

    Send stretch/move alerts during dip hours.

    Schedule group step challenges around peak activity windows.

#### Chart - 8: Number of Active Minutes Per Hour(Minute Level Plot)

```python
# Chart - 8 visualization code
# Filter for active minutes (non-zero steps)
active_minutes = minute_df[minute_df['TotalSteps'] > 0]

# Count active minutes per hour
active_minute_count = active_minutes.groupby('Hour').size().reset_index(name='ActiveMinutes')

# Plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=active_minute_count, x='Hour', y='ActiveMinutes', marker='o', color='seagreen')
plt.title('Number of Active Minutes Per Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Active Minutes')
plt.tight_layout()
plt.show()
```

##### 1. Why did you pick the specific chart?

To track how many minutes per hour users are actually active (i.e., not sitting still).

##### 2. What is/are the insight(s) found from the chart?

Only a few minutes per hour are active, suggesting opportunity for more consistent movement.

##### 3. Will the gained insights help creating a positive business impact?


Introduce ‚Äú10 active minutes per hour‚Äù goals.

Design progress badges for consistency (e.g., ‚ÄúYou moved every hour today!‚Äù)

#### Chart - 9: Average Daily Minutes by Activity Category(Daily Level Plot)

```python
# Chart - 9 visualization code

# Donut Chart for Average Minutes
avg_minutes = daily_df[['VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes', 'SedentaryMinutes']].mean()

plt.figure(figsize=(8, 6))
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
plt.pie(avg_minutes, labels=avg_minutes.index, autopct='%1.1f%%', startangle=140,
        colors=colors, wedgeprops={'width': 0.4})
plt.title('Average Daily Minutes by Activity Category (Donut Chart)')
plt.axis('equal')
plt.show()


```

##### 1. Why did you pick the specific chart?

A donut chart effectively communicates the proportion of time spent in different activity categories in a visually intuitive and engaging way.

##### 2. What is/are the insight(s) found from the chart?

The largest share of time is spent in SedentaryMinutes, followed by LightlyActiveMinutes. Very few minutes are spent in VeryActive or FairlyActive categories. This indicates a predominantly sedentary lifestyle for most users.

##### 3. Will the gained insights help creating a positive business impact?


Yes. These insights help:

    Design more engaging workout challenges in mobile health apps.

    Trigger sedentary alerts or reminders to stand or move.

    Offer personalized fitness programs encouraging increased Very Active minutes, which are more beneficial to health.

#### Chart - 10: Average Daily Distance by Activity Type(Daily Level Plot)

```python
# Chart - 10 visualization code
# Calculate average distances
avg_distances = daily_df[['VeryActiveDistance', 'ModeratelyActiveDistance', 'LightActiveDistance', 'SedentaryActiveDistance']].mean()

# Pie Chart
plt.figure(figsize=(8, 6))
colors = ['#c2c2f0', '#ffb3e6', '#c4e17f', '#ffcc99']
plt.pie(avg_distances, labels=avg_distances.index, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Average Daily Distance by Activity Type (Pie Chart)')
plt.axis('equal')
plt.show()
```

##### 1. Why did you pick the specific chart?

A pie chart allows a quick comparison of distances by activity level, suitable when you want to see contribution by parts to a whole.

##### 2. What is/are the insight(s) found from the chart?

LightActiveDistance typically dominates, suggesting most movement comes from light-intensity activities.

VeryActiveDistance is much lower, revealing fewer instances of intense physical activity.

##### 3. Will the gained insights help creating a positive business impact?


Yes. Understanding how much of users' movement comes from low vs. high intensity can:

    Inform goal setting in fitness tracking apps.

    Help healthcare providers identify low-activity users and offer targeted advice.

    Assist product managers in building features like distance streaks or step-based rewards to boost engagement.
